# Santigrattan Fahrenhayta dönüşüm yapan basit bir Yapay Sinir Ağı Modeli

Santigrattan Fahrenhayt'a dönüşümün formülü "f=c×1.8+32" biçimindedir. Şimdi çok basit bir yapay sinir ağına birkaç veri vererek bu denklemi bizim için bulmasını sağlayacağız.

## Kullanacağımız kütüphanelerin içe aktarılması

```
import tensorflow as tf
import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

## Eğitim setinin hazırlanması

```
celsius_q    = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
fahrenheit_a = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)
```

Burada modelimiz çok basit olacağı için çok fazla veriye ihtiyacımız yok. Modelin düzeyi arttıkça gereken veri sayısı da artar.

## Modelin hazırlanması

```
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

## Kayıp ve iyileştirme fonksiyonlarını da yazarak modelin tamamlanması

```
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))
```

## Modelin eğitilmesi

```
history = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=True)
```

## Eğitim istatistiklerinin görüntülenmesi

```
import matplotlib.pyplot as plt
plt.xlabel('Çağ Numarası')
plt.ylabel("Kayıp Büyüklüğü")
plt.plot(history.history['loss'])
```

## Modelin test edilmesi

```
print(model.predict([100.0]))
```
